### Desafio 1

Escreva um programa em Python para contabilizar a quantidade de ocorrências de cada palavra.

```python
palavras = [
   'red', 'green', 'black', 'pink', 'black', 'white', 'black', 'eyes',
   'white', 'black', 'orange', 'pink', 'pink', 'red', 'red', 'white', 'orange',
   'white', "black", 'pink', 'green', 'green', 'pink', 'green', 'pink',
   'white', 'orange', "orange", 'red'
]


# Seu código
import numpy as np
import pandas as pd
```

```python
palavras
```

```python
PalavrasUnicas = set(palavras)
PalavrasUnicas
```

```python
#lista
soma = []
def contando(PalavrasUnicas, palavras):
    for PalavrasUnicas in palavras:
        concatenar = PalavrasUnicas + ": " + str(palavras.count(PalavrasUnicas))
        soma.append(concatenar)
    return soma
```

```python
repetidas = list(set(contando(PalavrasUnicas, palavras)))
repetidas
```

```python
#dicionario
soma = []
def ContandoDic(PalavrasUnicas, palavras):
    for PalavrasUnicas in palavras:
        concatenar = str(palavras.count(PalavrasUnicas))
        soma.append(concatenar)
    return soma
```

```python
RepetidasDic = ContandoDic(PalavrasUnicas,palavras)
```

```python
PalavrasDicionario = dict(zip(palavras,RepetidasDic))
```
