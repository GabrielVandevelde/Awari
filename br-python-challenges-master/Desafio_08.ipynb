### Desafio 8

Escreva um script Python para encontrar as 10 palavras mais longas em um arquivo.

O arquivo TXT está localizado na mesma pasta do projeto (texto.txt).

```python
# Define lista para tamanho das palavras
GrWords = []

#Comando para abrir o arquivo
with open('texto.txt') as file:
    data=file.read().split() # Transforma o arquivo em lista
       
    for i in range(len(data)): # Preenche a lista com o tamanho das palavras do arquivo
        GrWords.append(len(data[i]))
    
    GrWords = sorted(set(GrWords), reverse=True) # Ordena a lista do maior para o menor sem repetições
    print ()

    for i in range(0,10): # Retorna o numero de caracteres das dez maires palavras
        res=[word for word in data if len(word)==GrWords[i]]  
        ct = i+1
        print(str(ct)+" - Quantidade de letras", GrWords[i] , res)
```
